```python
class Manager():
    def __init__(self):
        self.accounts = {}
        self.index = 0
        
    def check_index(self):
        index = int(input('What is your ID: '))
        try:
            index = int(index)
        except ValueError:
            raise ExceptionHandling('Please, write an int value')
        if index not in self.accounts:
            raise ExceptionHandling('You entered incorrect index')
        return index
    
    def create_account(self):
        print('**Creating new account**')
        name = str(input('What is your name: '))
        password = str(input('What is your password: '))
        starting_amount = int(input('What is your starting amount: '))
        Account = Bank(name, password, starting_amount)
        Account.check_balance(starting_amount)
        print('Your id is: {}'.format(self.index))
        self.accounts[self.index] = Account
        self.index +=1
        print('\n')
        print('Account has been created successfully!')
        print('\n')
    
    def delete_account(self):
        print('**Delete account**')
        password = str(input('What is your password: '))
        index = self.check_index()
        account = self.accounts[index]
        password = account.check_password(password)
        if account.balance != 0:
            print('\n')
            print('First, withdraw all of your money')
            print('\n')
        else:
            print('\n')
            print('Your account has been deleted')
            print('\n')
            del self.accounts[index]
        
    def deposit(self):
        print('**Deposit**')
        password = str(input('What is your password: '))
        index = self.check_index()
        deposit_amount = int(input('What is your deposit amount: '))
        account = self.accounts[index]
        account.check_balance(deposit_amount)
        account.check_password(password)
        print('\n')
        print(account.deposit(password, deposit_amount))
        print('\n')
    
    def withdraw(self):
        print('**Withdraw**')
        index = self.check_index()
        password = str(input('What is your password: '))
        withdraw_amount = int(input('What is your withdraw amount: '))
        account = self.accounts[index]
        account.check_withdrawl(withdraw_amount)
        account.check_password(password)
        print('\n')
        print(account.withdrawl(password, withdraw_amount)) 
        print('\n')
        
              
    def balance(self):
        print('**Balance**')
        index = self.check_index()
        password = str(input('What is your password: '))
        account = self.accounts[index] 
        account.check_password(password)
        print('\n')
        print(account.getbalance(password))
        print('\n')
        
             
```
